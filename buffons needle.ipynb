```python
import numpy as np
import matplotlib.pyplot as plt

ntot = int(input("n total ")) #number of throwing
#d > l
d = float(input("d ")) #distance between line
l = float(input("l ")) #length of needle
```

```python
#searching value of pi
def needle():
    nhit = 0
    for i in range(0, ntot):
        theta = np.random.uniform(0, np.pi/2)
        x = np.random.uniform(0,d)
        if (x <= l*np.cos(theta)):
            nhit = nhit + 1
    phi = 2*l*ntot/(d*nhit)
    return phi
```

```python
exp = 1000 #number of experiment
pi_arr = []
for i in range(0, exp):
    phi = needle()
    pi_arr.append(phi)
    print(pi_arr[i])
```

```python
#value of pi
pi_avg = sum(pi_arr) / exp
print (pi_avg)
```

```python
#make plot for phi
plt.plot(pi_arr, c='green')
plt.axhline(y = pi_avg, c = 'black')
plt.show()
```
